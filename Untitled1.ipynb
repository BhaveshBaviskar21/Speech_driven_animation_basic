```python
def get_positional_encoding(max_len, d_model):
    pos_enc = np.zeros((max_len, d_model))
    for pos in range(max_len):
        for i in range(d_model):
            if i % 2 == 0:
                pos_enc[pos, i] = np.sin(pos / (10000 ** (i / d_model)))
                # print("i%2 == 0 : "+str(pos)+" : "+str(i)+" : "+str(np.sin(pos / (10000 ** (i / d_model)))))
            else:
                pos_enc[pos, i] = np.cos(pos / (10000 ** ((i - 1) / d_model)))
                # print("i%2 != 0 :"+str(pos)+" : "+str(i)+" : "+ str(np.cos(pos / (10000 ** ((i -1) / d_model)))))
    return pos_enc
```

```python
class DCGAN(Model):
    def __init__(self, generator, discriminator):
        super().__init__()
        self.generator = generator
        self.discriminator = discriminator
        self.g_loss_metric = Mean(name='g_loss')
        self.d_loss_metric = Mean(name='d_loss')

    @property
    def metrics(self):
        return [self.g_loss_metric, self.d_loss_metric]

    def compile(self, g_optimizer, d_optimizer, loss_fn):
        super(DCGAN, self).compile()
        self.g_optimizer = g_optimizer
        self.d_optimizer = d_optimizer
        self.loss_fn = loss_fn

    def fit(self,steps):

        start = time.time()
        for step in range(steps):
            for f in ["fadg0","faks0","fcft0","fcmh0","fcmr0","fcrh0"]:
                fake_data = np.load("/content/drive/MyDrive/ml_project/pre_proc_data/fake_data_"+f+".npy")
                real_data = np.load("/content/drive/MyDrive/ml_project/pre_proc_data/real_data_"+f+".npy")
                print(f"Step: {step+1}")
                for i in range(len(real_data)):
                    self.train_step(real_images=real_data[i][np.newaxis,:],fake_data=fake_data[i][np.newaxis,:])
            print(f'Time taken for step: {time.time()-start:.2f} sec')
            start=time.time()

    @tf.function
    def train_step(self, real_images,fake_data):

        # batch_size = tf.shape(real_images)[0]

        # train the discriminator with real (1) and fake (0) images
        with tf.GradientTape() as d_tape:
            # compute loss on real images
            pred_real = self.discriminator([fake_data,real_images], training=True)
            # generate real image labels
            real_labels = tf.ones_like(pred_real)
            # label smoothing
            real_labels += 0.02 * tf.random.uniform(tf.shape(real_labels))
            d_loss_real = self.loss_fn(real_labels, pred_real)

            # compute loss on fake images
            fake_images = self.generator(fake_data,training=False)
            pred_fake = self.discriminator([fake_data,fake_images], training=True)
            # generate fake labels
            fake_labels = tf.zeros_like(pred_fake)
            d_loss_fake = self.loss_fn(fake_labels, pred_fake)

            # total discriminator loss
            d_loss = (d_loss_real + d_loss_fake) / 2

        # compute discriminator gradients
        gradients = d_tape.gradient(d_loss, self.discriminator.trainable_variables)
        # update the gradients
        self.d_optimizer.apply_gradients(zip(gradients, self.discriminator.trainable_variables))


        # generator want discriminator to think that fake images are real
        with tf.GradientTape() as g_tape:
            # generate fake images from generator
            fake_images = self.generator(fake_data, training=True)
            # classify images as real or fake
            pred_fake = self.discriminator([fake_data,fake_images], training=True)
            # compute loss
            g_loss = self.loss_fn(tf.ones_like(pred_fake), pred_fake)

        # compute gradients
        gradients = g_tape.gradient(g_loss, self.generator.trainable_variables)
        # update the gradients
        self.g_optimizer.apply_gradients(zip(gradients, self.generator.trainable_variables))

        # update states for both models
        self.d_loss_metric.update_state(d_loss)
        self.g_loss_metric.update_state(g_loss)

        return {'d_loss': self.d_loss_metric.result(), 'g_loss': self.g_loss_metric.result()}

```
